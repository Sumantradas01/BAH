```python
def load_dataset(name):

    info = DATASETS[name]

    collection = (
        ee.ImageCollection(info["id"])
        .filterDate(START_DATE, END_DATE)
        .select(info["band"])
    )

    return collection.mean().clip(INDIA)
```
